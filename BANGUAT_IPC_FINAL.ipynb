```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import urllib
import re
```

```python
url = "https://banguat.gob.gt/page/base-diciembre-de-20101000"  
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
links = soup.find_all("a", href=True)

file_urls = []

for link in links:
    href = link["href"]
    if re.search(r"\.(xls|xlsx|csv)$", href):
        file_urls.append(href)

# Base URL (domain name)
base_url = 'https://banguat.gob.gt/'

full_url = urllib.parse.urljoin(base_url, file_urls[0]) 
print(full_url)
print(full_url == "https://banguat.gob.gt/sites/default/files/banguat/estaeco/SR/sr005_1.xls")

df = pd.read_excel(full_url, skiprows=5,names=["Periodo","Indice","Variacion Intermensual","Variacion Internanual","Variacion Acumulada","Variación Promedio"],usecols="C:H")
df["Year"] = np.where(df["Indice"].isna()==True,df["Periodo"],np.nan )
df["Year"] = df["Year"].ffill(axis=0)
df_final = df.dropna().reset_index(drop=True) 

dict_mes = {
    "Enero":1,
    "Febrero":2,
    "Marzo":3,
    "Abril":4,
    "Mayo":5,
    "Junio":6,
    "Julio":7,
    "Agosto":8,
    "Septiembre":9,
    "Octubre":10,
    "Noviembre":11,
    "Diciembre":12


}
df_final["Month"] = df_final["Periodo"].map(dict_mes)
df_final["Day"] =1
df_final['Date'] = pd.to_datetime(df_final[['Year', 'Month', 'Day']])
df_final = df_final[["Date","Indice","Variacion Intermensual","Variacion Internanual","Variacion Acumulada","Variación Promedio"]]
df_final

```
